```python
import numpy as np, pandas as pd
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.cross_validation import cross_val_score
```

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
y = train['Response'].values
id_test = test['Id'].values
piv_train = train.shape[0]
df_all = pd.concat((train,test), axis=0, ignore_index=True).drop(['Id','Response'], axis=1)
#One-hot encoding for product into 2 only
pi2_dummies = pd.get_dummies(train['Product_Info_2'], prefix='Product_Info_2')
df_all = df_all.drop(['Product_Info_2'], axis=1)
df_all = pd.concat((df_all, pi2_dummies), axis=1)
df_all.fillna(-1, inplace=True)
vals = df_all.values
X = vals[:piv_train]
X_test = vals[piv_train:]
```

```python
classifier = RandomForestClassifier(random_state=0, n_estimators=100)
#score = cross_val_score(estimator, X, risk).mean()
#print("Score with the entire dataset = %.2f" % score)
classifier.fit(X,y)
classifier.score(X,y)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
#Generate submission
sub = pd.DataFrame(np.column_stack((id_test, y_pred)), columns=['Id', 'Response'])
sub.head(50)
```

```python
sub.to_csv('../output/sub.csv',index=False)
```
